```python
#Cuento la cantidad de géneros por juego que hay y creo un ranking de géneros.

genres = df["genres"].str.split(',').explode().str.strip()
genres = genres.str.replace(r"[\[\]']", '', regex=True)
genre_counts = genres.value_counts().reset_index()
genre_counts.columns = ["genres", "count"]
genre_counts["position"] = range(1, len(genre_counts) + 1)
```
